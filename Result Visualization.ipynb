```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data into a DataFrame (fill in the exact data from your table)
data = {
    'Model': ['Model 1 Before', 'Model 1 Pruning', 'Model 1 Pruning+Quant',
              'Model 2 Before', 'Model 2 Pruning', 'Model 2 Pruning+Quant',
              'GhostNet Before', 'GhostNet Pruning', 'GhostNet Pruning+Quant',
              'AO-CNN Before', 'AO-CNN Pruning', 'AO-CNN Pruning+Quant'],
    'Model Size (KB)': [43.73, 7.01, 6.25, 1473, 479, 242, 380, 62.91, 41.38, 43.73, 7.01, 6.25],
    'Training Time (s)': [185, 152, 153, 2255, 227, 204, 754, 571, 543, 50.85, 49.15, 80.67],
    'Training Accuracy (%)': [94.14, 95.53, 95.89, 99.82, 99.89, 99.89, 99.13, 99.61, 99.61, 95.22, 93.97, 95.52],
    'Testing Accuracy (%)': [94.73, 96.13, 96.29, 99.15, 99.02, 99.13, 98.51, 98.68, 98.76, 95.25, 93.97, 95.52]
}

df = pd.DataFrame(data)

# Set the plot style
sns.set(style="whitegrid")

# Plot training time comparison
plt.figure(figsize=(14, 6))
sns.barplot(x='Model', y='Training Time (s)', data=df, palette="viridis")
plt.title('Training Time Comparison')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot training accuracy comparison
plt.figure(figsize=(14, 6))
sns.barplot(x='Model', y='Training Accuracy (%)', data=df, palette="viridis")
plt.title('Training Accuracy Comparison')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot testing accuracy comparison
plt.figure(figsize=(14, 6))
sns.barplot(x='Model', y='Testing Accuracy (%)', data=df, palette="viridis")
plt.title('Testing Accuracy Comparison')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Plot model size comparison with logarithmic scale
plt.figure(figsize=(14, 6))
sns.barplot(x='Model', y='Model Size (KB)', data=df, palette="viridis")
plt.title('Model Size Comparison (Log Scale)')
plt.xticks(rotation=45, ha='right')
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.ylabel('Model Size (KB) (Log Scale)')  # Update the y-axis label
plt.tight_layout()
plt.show()
```
